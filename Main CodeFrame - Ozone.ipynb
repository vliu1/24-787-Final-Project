```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Ridge
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error
```

```python
dataframe = pd.read_csv('Test3.csv')
data = np.array(dataframe)
```

```python
column_heads = list(dataframe.columns.values)
FIPS = data[:,0]
X = data[:,5:]
Y = data[:,3]

scaler = StandardScaler()
scaler.fit(X)
xPrime = scaler.transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(xPrime, Y, test_size=0.3)

# y_train = np.reshape(y_train, (-1, 1))
# y_test = np.reshape(y_test, (-1, 1))
# X_train = np.reshape(X_train, (-1, 1))
# X_test = np.reshape(X_test, (-1, 1))

LR = LinearRegression().fit(X_train, y_train)
    

LR_y_train_pred = LR.predict(X_train)
LR_y_test_pred = LR.predict(X_test)
    
    

LR_train_score = r2_score(y_train, LR_y_train_pred)
LR_test_score = r2_score(y_test, LR_y_test_pred)

print(LR)
print('scores: train = ',np.around(LR_train_score,2),', test = ',np.around(LR_test_score,2))
```

```python
# Linear Regression unnormalized
LR_train_score = 0
LR_test_score = 0
Ridge_train_score = 0
Ridge_test_score = 0
SVR_train_score = 0
SVR_test_score = 0

for i in range(1000):
    
    X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3)


    LR = LinearRegression().fit(X_train, y_train)
    ridge = Ridge(alpha=1).fit(X_train,y_train)
    svr = SVR(C=1.0, epsilon=0.1).fit(X_train,y_train)
    

    LR_y_train_pred = LR.predict(X_train)
    LR_y_test_pred = LR.predict(X_test)

    Ridge_y_train_pred = ridge.predict(X_train)
    Ridge_y_test_pred = ridge.predict(X_test)

    SVR_y_train_pred = svr.predict(X_train)
    SVR_y_test_pred = svr.predict(X_test)
    
    


    LR_train_score1 = r2_score(y_train, LR_y_train_pred)
    LR_test_score1 = r2_score(y_test, LR_y_test_pred)

    Ridge_train_score1 = r2_score(y_train, Ridge_y_train_pred)
    Ridge_test_score1 = r2_score(y_test, Ridge_y_test_pred)

    SVR_train_score1 = r2_score(y_train, SVR_y_train_pred)
    SVR_test_score1 = r2_score(y_test, SVR_y_test_pred)




    if LR_train_score1>LR_train_score:
        LR_train_score = LR_train_score1
    if LR_test_score1>LR_test_score:
        LR_test_score = LR_test_score1

    if Ridge_train_score1>Ridge_train_score:
        Ridge_train_score = Ridge_train_score1
    if Ridge_test_score1>Ridge_test_score:
        Ridge_test_score = Ridge_test_score1

    if SVR_train_score1>SVR_train_score:
        SVR_train_score = SVR_train_score1
    if SVR_test_score1>SVR_test_score:
        SVR_test_score = SVR_test_score1


print(LR)
print('scores: train = ',np.around(LR_train_score,2),', test = ',np.around(LR_test_score,2),'\n')

print(ridge)
print('scores: train = ',np.around(Ridge_train_score,2),', test = ',np.around(Ridge_test_score,2),'\n')

print(svr)
print('scores: train = ',np.around(SVR_train_score,2),', test = ',np.around(SVR_test_score,2),'\n')
```

```python
# Linear Regression normalized
LR_train_score = 0
LR_test_score = 0
Ridge_train_score = 0
Ridge_test_score = 0
SVR_train_score = 0
SVR_test_score = 0

k = 5

for i in range(k):
    
    X_train, X_test, y_train, y_test = train_test_split(xPrime, Y, test_size=0.3)


    LR = LinearRegression().fit(X_train, y_train)
    ridge = Ridge(alpha=1).fit(X_train,y_train)
    svr = SVR(C=1.0, epsilon=0.1).fit(X_train,y_train)
    

    LR_y_train_pred = LR.predict(X_train)
    LR_y_test_pred = LR.predict(X_test)

    Ridge_y_train_pred = ridge.predict(X_train)
    Ridge_y_test_pred = ridge.predict(X_test)

    SVR_y_train_pred = svr.predict(X_train)
    SVR_y_test_pred = svr.predict(X_test)
    
    


    LR_train_score1 = mean_squared_error(y_train, LR_y_train_pred)
    LR_test_score1 = mean_squared_error(y_test, LR_y_test_pred)

    Ridge_train_score1 = mean_squared_error(y_train, Ridge_y_train_pred)
    Ridge_test_score1 = mean_squared_error(y_test, Ridge_y_test_pred)

    SVR_train_score1 = mean_squared_error(y_train, SVR_y_train_pred)
    SVR_test_score1 = mean_squared_error(y_test, SVR_y_test_pred)
    
    LR_train_score += LR_train_score1
    LR_test_score += LR_test_score1
    Ridge_train_score += Ridge_train_score1
    Ridge_test_score += Ridge_test_score1
    SVR_train_score += SVR_train_score1
    SVR_test_score += SVR_test_score1

    
LR_train_score = LR_train_score/k
LR_test_score = LR_test_score/k
Ridge_train_score = Ridge_train_score/k
Ridge_test_score = Ridge_test_score/k
SVR_train_score = SVR_train_score/k
SVR_test_score = SVR_test_score/k
    


print(LR)
print('scores: train = ',np.around(LR_train_score,6),', test = ',np.around(LR_test_score,6),'\n')

print(ridge)
print('scores: train = ',np.around(Ridge_train_score,6),', test = ',np.around(Ridge_test_score,6),'\n')

print(svr)
print('scores: train = ',np.around(SVR_train_score,6),', test = ',np.around(SVR_test_score,6),'\n')
```
